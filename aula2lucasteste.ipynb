```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
#Leitura do Dataset
data = pd.read_csv('gas_price.csv')
```

```python
data.head(10)
```

```python
# Quantidade de Postos Pesquisados por Região (Gráfico em barras)
data['REGIÃO'].value_counts().plot(kind='bar', figsize=(11,5), grid=False, rot=0, color='blue')
plt.title('Quantidade de Postos Pesquisados por Regiao')
plt.xlabel('Regioes')
plt.ylabel('Qtd. de Postos')
plt.show()
```

```python
# Quantidade de Postos Pesquisados por Região
data['REGIÃO'].value_counts().head(10).plot(kind='barh', figsize=(11,5), grid=False, rot=0, color='blue')
plt.title('Quantidade de Postos Pesquisados por Regiao')
plt.xlabel('Qtd. de Postos')
plt.ylabel('Regioes')
plt.show()
```

```python
# Quantidade de Postos Pesquisados por Região
data['REGIÃO'].value_counts().head(10).plot(kind='pie', figsize=(8,8), grid=False, rot=0)
plt.title('Quantidade de Postos Pesquisados por Regiao')
plt.xlabel('Regioes')
plt.ylabel('Regioes')
plt.show()
```
