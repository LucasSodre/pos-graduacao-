```python
# Importando Matplotlib e Numpy
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Importando o módulo de Regressão Linear do scikit-learn
from sklearn.linear_model import LinearRegression
```

```python
# Preparando os dados de treino

idade = [[20],[36],[43],[71],[18],[25],[34],[40],[81],[19],[27],[23]]

altura = [[170], [185], [179], [160], [181],[169], [177], [172], [150], [186],[179],[175]]

peso = [[58], [78], [100], [50], [70],[68], [76], [105], [56], [63],[64],[61]]

print("-----------------------------------------------")
print("peso")
print(len(peso))
print("altura")
print(len(altura))
print("idade")
print(len(idade))
print("-----------------------------------------------")
```

```python
# Criando o modelo 1
modelo = LinearRegression()

# Criando o modelo 2
modelo1 = LinearRegression()

# Criando o modelo 3
modelo2 = LinearRegression()
```

```python
# Scatter Plot representando a regressão linear
plt.scatter(altura, peso,  color = 'black')
plt.plot(altura, modelo.predict([[200],[150],[145],[170],[169]]), color = 'blue', linewidth = 3)
plt.xlabel('X')
plt.ylabel('Y')
plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# Coeficientes
print('Coeficiente: \n', modelo.coef_)

# MSE (mean square error)
print("MSE: %.2f" % np.mean((modelo.predict(peso) - altura) ** 2))

# Score de variação: 1 representa predição perfeita
print('Score de variação: %.2f' % modelo.score(peso, altura))
```

```python
# Coeficientes
print('Coeficiente: \n', modelo1.coef_)

# MSE (mean square error)
print("MSE: %.2f" % np.mean((modelo1.predict(idade) - altura) ** 2))

# Score de variação: 1 representa predição perfeita
print('Score de variação: %.2f' % modelo1.score(idade, altura))
```

```python
# Coeficientes
print('Coeficiente: \n', modelo2.coef_)

# MSE (mean square error) mais proximo de 1
print("MSE: %.2f" % np.mean((modelo2.predict(idade) - peso) ** 2))

# Score de variação: 1 representa predição perfeita menos proximo de 1
print('Score de variação: %.2f' % modelo2.score(idade, peso))
```
